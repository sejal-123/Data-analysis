```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,4,11,14,20]
plt.plot(x,y)
```

```python
x = [1,2,3,4,5]
y = [2,4,11,14,20]
plt.plot(x,y,color='red',linestyle='-')
```

```python
x = [1,2,3,4,5]
y = [2,4,11,14,20]
plt.plot(x,y,color='red',linestyle='--')
```
